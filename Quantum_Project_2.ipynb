```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
```

```python
#hbar = 1.054571817E-34
hbar=1
#m = 9.1093837E-31
m = 1
#L = 10E-9
L = 10
N = 10
d = L/N
const = -hbar**2/2/m/d**2
x = np.linspace(0,L,N)
```

```python
#Building the matrix and calculating the eigenvalues and eigenvectors
H = []

for i in np.arange(N):
    H.append(-2)

H = np.diag(H)
H = np.array(H)

H[np.triu_indices(N, k=1)] = 1   # Set the "upper triangle" elements to 1
H[np.tril_indices(N, k=-1)] = 1  # Set the "lower triangle" elements to 1
H[np.triu_indices(N, k=2)] = 0   # Set the "more upper triangle" elements to 0
H[np.tril_indices(N, k=-2)] = 0  # Set the "more lower triangle" elements to 0

H = const*H

eigen = np.linalg.eig(H)
```

```python
# N=10
n=5
eigs = []
for i in np.arange(N):
    eig = [eigen[0][i], np.ndarray.flatten(eigen[1][:,i])]
    eigs.append(eig)

eigs = np.array(eigs)
eigs = eigs[eigs[:,0].argsort()]
mineigs = eigs[0:n]

E = []
for i in np.arange(0,n):
    E.append(((i+1)*np.pi*hbar)**2/2/m/L**2)
    print(mineigs[i][0])
print(E)

plt.figure(figsize=(12, 10), dpi=80)

for i in np.arange(1,6):
    
    plt.subplot(5,2,2*i-1)
    plt.plot(np.linspace(0,L,1000),np.sqrt(2/L)*np.sin(i*np.pi*(np.linspace(0,L,1000))/L))
    plt.plot(x,mineigs[i-1][1])
    
    plt.subplot(5,2,2*i)
    plt.plot(np.linspace(0,L,1000), (np.sqrt(2/L)*np.sin(i*np.pi*(np.linspace(0,L,1000))/L))**2)
    plt.plot(x,mineigs[i-1][1]**2)
```
